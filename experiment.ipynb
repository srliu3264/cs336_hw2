```python
s = torch.tensor(0,dtype=torch.float32)
for i in range(1000):
    s += torch.tensor(0.01,dtype=torch.float32)
print(s)
```

```python
s = torch.tensor(0,dtype=torch.float16)
for i in range(1000):
    s += torch.tensor(0.01,dtype=torch.float16)
print(s)
```

```python
s = torch.tensor(0,dtype=torch.float32)
for i in range(1000):
    s += torch.tensor(0.01,dtype=torch.float16)
print(s)
```

```python
s = torch.tensor(0,dtype=torch.float32)
for i in range(1000):
    x = torch.tensor(0.01,dtype=torch.float16)
    s += x.type(torch.float32)
print(s)
```

```python
import pandas as pd
import numpy as np
```

```python
compiled = pd.read_csv("./results/attn_bench/attn_compiled.csv")
uncompiled = pd.read_csv("./results/attn_bench/attn_uncompiled.csv")
```

```python
compare = pd.merge(left=uncompiled,right=compiled[["d","T","fwd_ms","mem_before_bwd_mib","bwd_ms","peak_mib","status"]],on=["d","T"], how = "left",suffixes=("_uncom", "_comp"))
```

```python
compare.head(5)
```

```python
result = compare[["d","T", "fwd_ms_uncom","fwd_ms_comp","bwd_ms_uncom","bwd_ms_comp"]]
```

```python
result["speedup_bwd"] = (result["bwd_ms_uncom"]/result["bwd_ms_comp"])
```

```python
result.head(5)
```

```python
print(result.to_latex(index=False,float_format="%.2f", caption="orch.compile vs. uncompiled attention",label="table:compare_attn"))
```

```python
xl_compiled = pd.read_csv("./results/compile_compare/compile_xl_compiled.csv")
xl_un = pd.read_csv("./results/compile_compare/compile_xl.csv")
large_compiled = pd.read_csv("./results/compile_compare/compile_large_compiled.csv")
large_un = pd.read_csv("./results/compile_compare/compile_large.csv")
medium_compared = pd.read_csv("./results/compile_compare/compile_medium_compiled.csv")
medium_un = pd.read_csv("./results/compile_compare/compile_medium.csv")
```

```python
small_compiled = pd.read_csv("./results/compile_compare/compile_small_compiled.csv")
small_un = pd.read_csv("./results/compile_compare/compile_small.csv")
```

```python
compiled = pd.concat([small_compiled, medium_compared, large_compiled,xl_compiled])
un = pd.concat([small_un, medium_un, large_un, xl_un])
```

```python
compiled.head(5)
```

```python
compare = pd.merge(left = un[["size","n_params","mean_ms","std_ms","max_ms","min_ms"]], right = compiled[["size","mean_ms","std_ms","max_ms","min_ms"]], how = "left", on = "size", suffixes=("_un","_comp"))
```

```python
compare.head(5)
```

```python
print(compare.to_latex(index=False, float_format = "%.2f", caption="Compare run time: un-compiled vs JIT compiled (batch size 4. warmup 10, steps 10)", label="table: compare_compile_jit"))
```

```python
forward_xl_un = pd.read_csv("./results/compile_compare/forward_only/forward_compile_xl.csv")
forward_xl_comp = pd.read_csv("./results/compile_compare/forward_only/forward_compile_xl_compiled.csv")
```

```python
forward_large_un = pd.read_csv("./results/compile_compare/forward_only/forward_compile_large.csv")
forward_large_comp = pd.read_csv("./results/compile_compare/forward_only/forward_compile_large_compiled.csv")
```

```python
forward_med_un = pd.read_csv("./results/compile_compare/forward_only/forward_compile_medium.csv")
forward_med_comp = pd.read_csv("./results/compile_compare/forward_only/forward_compile_medium_compiled.csv")
```

```python
forward_small_un = pd.read_csv("./results/compile_compare/forward_only/forward_compile_small.csv")
forward_small_comp = pd.read_csv("./results/compile_compare/forward_only/forward_compile_small_compiled.csv")
```

```python
forward_un = pd.concat([forward_small_un,forward_med_un,forward_large_un,forward_xl_un])
forward_comp = pd.concat([forward_small_comp,forward_med_comp,forward_large_comp,forward_xl_comp])
```

```python
forward_compare = pd.merge(left = forward_un[["size","n_params","mean_ms","std_ms","max_ms","min_ms"]], right = forward_comp[["size","mean_ms","std_ms","max_ms","min_ms"]], how = "left", on = "size", suffixes=("_un","_comp"))
```

```python
print(forward_compare.to_latex(index=False, float_format = "%.2f", caption="Compare run time (forward): un-compiled vs JIT compiled (batch size 4. warmup 10, steps 10)", label="table: fwd_compare_compile_jit"))
```
