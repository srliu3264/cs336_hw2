```python
s = torch.tensor(0,dtype=torch.float32)
for i in range(1000):
    s += torch.tensor(0.01,dtype=torch.float32)
print(s)
```

```python
s = torch.tensor(0,dtype=torch.float16)
for i in range(1000):
    s += torch.tensor(0.01,dtype=torch.float16)
print(s)
```

```python
s = torch.tensor(0,dtype=torch.float32)
for i in range(1000):
    s += torch.tensor(0.01,dtype=torch.float16)
print(s)
```

```python
s = torch.tensor(0,dtype=torch.float32)
for i in range(1000):
    x = torch.tensor(0.01,dtype=torch.float16)
    s += x.type(torch.float32)
print(s)
```

```python
import pandas as pd
import numpy as np
```

```python
compiled = pd.read_csv("./results/attn_bench/attn_compiled.csv")
uncompiled = pd.read_csv("./results/attn_bench/attn_uncompiled.csv")
```

```python
compare = pd.merge(left=uncompiled,right=compiled[["d","T","fwd_ms","mem_before_bwd_mib","bwd_ms","peak_mib","status"]],on=["d","T"], how = "left",suffixes=("_uncom", "_comp"))
```

```python
compare.head(5)
```

```python
result = compare[["d","T", "fwd_ms_uncom","fwd_ms_comp","bwd_ms_uncom","bwd_ms_comp"]]
```

```python
result["speedup_bwd"] = (result["bwd_ms_uncom"]/result["bwd_ms_comp"])
```

```python
result.head(5)
```

```python
print(result.to_latex(index=False,float_format="%.2f", caption="orch.compile vs. uncompiled attention",label="table:compare_attn"))
```
